Next we will want to process all of the images. 
The process involves (step-1)removing distortion; (step-2) registeration and (step-3) rectification.
>(Step 1)
    To remove distortion we have to include the site-specific SiteDistortion class object, these are pre-loaded when you import scripts.undistort_tools.py 
    
>(Step 2)
     To register the image we need to specify the path to the reference image, to which these images are being registered
     
>(Step 3)
    To rectify the image we need to specify the site- and date-specific PanelDate object. These were loaded with we imported scripts.rectification_tools.py
    
**These steps will all happen in sequence together, but we must first define our 'ref_image_path', 'SiteDistortion' class object, and our 'PanelDate' class object.**
